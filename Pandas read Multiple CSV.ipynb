```python
stocks = ['bbl', 'ktb', 'scb']
dfs = []

location_prefix = 'https://github.com/prasertcbs/tutorial/raw/master/eod/'
dfs = [pd.read_csv(f'{location_prefix}{s}.csv', parse_dates=True, index_col='Date') for s in stocks]


```
