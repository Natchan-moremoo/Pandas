# ตั้งชื่อคอลัมม์เอง

```python
colnames = ['seq', 'brand', '1300cc', '1600cc', '1800cc', '2000cc', 'gt2000cc', 'elec', 'total']
```

```python
df = pd.read_excel(url, sheet_name='Total', 
                   skiprows=7, skipfooter=3,
                  header=None, names=colnames)
df
```

```python
df.drop('seq', axis=1, inplace=True)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.set_index('brand', inplace=True)
df
```

```python
df = df[df.total != 0]
df
```

```python
df.nlargest(10, 'total')
```

```python
df.nlargest(10, 'total')['total'].plot(kind='barh')
```

```python
df.nlargest(10, 'total')['total'].sort_values().plot(kind='barh')
```
