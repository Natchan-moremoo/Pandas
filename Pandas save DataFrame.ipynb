# Save DataFrame to CSV

```python
import pandas as pd
print(f'pandas version: {pd.__version__}')
```

```python
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/BaskinRobbins.csv'
df = pd.read_csv(url)
df
```

```python
df.to_csv('Pandas/Data/test_save1.csv')
```

```python
!cat Pandas/Data/test_save1.csv
# window cat เปลี่ยนเป็น type
```

# เพิ่มชื่อหัว index

```python
df.to_csv('Pandas/Data/test_save2.csv', index_label='id')
```

```python
!cat Pandas/Data/test_save2.csv
```

# ตัดคอลัมน์ index ออก

```python
df.to_csv('Pandas/Data/test_save3.csv', index=False)
```

```python
!cat Pandas/Data/test_save3.csv
```

# อยาก save บางคอลัมน์

```python
df.columns #เช็คมีกี่คอลัมน์
```

```python
df.to_csv('Pandas/Data/test_save4.csv', index=False, columns=['Flavour', 'Calories', 'Sugars (g)'])
```

```python
!cat Pandas/Data/test_save4.csv
```
